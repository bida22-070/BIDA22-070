```python
#Exercise 1 Temperature converter
def temperature_converter():
    """
    Convert temperature between Fahrenheit and Centigrade
    """
    print("Temperature Converter")
    print("1. Fahrenheit to Centigrade")
    print("2. Centigrade to Fahrenheit")

    try:
        choice = int(input("Enter your choice (1 or 2): "))

        if choice == 1:
            # Fahrenheit to Centigrade
            fahrenheit = float(input("Enter temperature in Fahrenheit: "))
            centigrade = (fahrenheit - 32) / 1.8
            print(f"{fahrenheit}°F = {centigrade:.2f}°C")

        elif choice == 2:
            # Centigrade to Fahrenheit
            centigrade = float(input("Enter temperature in Centigrade: "))
            fahrenheit = (centigrade * 1.8) + 32
            print(f"{centigrade}°C = {fahrenheit:.2f}°F")

        else:
            print("Invalid choice! Please enter 1 or 2.")

    except ValueError:
        print("Please enter valid numbers!")

# Run the converter
temperature_converter()

# Additional: Enhanced version with loop
def temperature_converter_enhanced():
    """
    Enhanced temperature converter with continuous operation
    """
    while True:
        print("\n" + "="*40)
        print("TEMPERATURE CONVERTER")
        print("="*40)
        print("1. Fahrenheit to Centigrade")
        print("2. Centigrade to Fahrenheit")
        print("3. Exit")

        try:
            choice = input("Enter your choice (1-3): ").strip()

            if choice == '1':
                fahrenheit = float(input("Enter temperature in Fahrenheit: "))
                centigrade = (fahrenheit - 32) / 1.8
                print(f"Result: {fahrenheit}°F = {centigrade:.2f}°C")

            elif choice == '2':
                centigrade = float(input("Enter temperature in Centigrade: "))
                fahrenheit = (centigrade * 1.8) + 32
                print(f"Result: {centigrade}°C = {fahrenheit:.2f}°F")

            elif choice == '3':
                print("Thank you for using Temperature Converter!")
                break

            else:
                print("Invalid choice! Please enter 1, 2, or 3.")

        except ValueError:
            print("Error: Please enter valid numbers!")
        except KeyboardInterrupt:
            print("\nProgram interrupted. Goodbye!")
            break

# Uncomment to run enhanced version
# temperature_converter_enhanced()
```

```python
#Exercise 2 Number Averager
def average_numbers():
    """
    Calculate average of any number of numbers
    """
    numbers = []

    print("Number Averager")
    print("Enter numbers one by one")
    print("Enter 'done' to calculate average")

    while True:
        user_input = input("Enter a number: ").strip().lower()

        if user_input == 'done':
            break

        try:
            number = float(user_input)
            numbers.append(number)
            print(f"Added: {number}")

        except ValueError:
            print("Please enter a valid number or 'done' to finish")

    if numbers:
        total = sum(numbers)
        count = len(numbers)
        average = total / count

        print(f"\nResults:")
        print(f"Numbers entered: {numbers}")
        print(f"Total numbers: {count}")
        print(f"Sum: {total}")
        print(f"Average: {average:.2f}")
    else:
        print("No numbers were entered!")

# Run the averager
average_numbers()

# Additional: Enhanced version with statistics
def advanced_averager():
    """
    Advanced number averager with additional statistics
    """
    numbers = []

    print("Advanced Number Analyzer")
    print("=" * 30)

    while True:
        print("\nOptions:")
        print("1. Add number")
        print("2. Calculate statistics")
        print("3. Reset list")
        print("4. Exit")

        choice = input("Choose option (1-4): ").strip()

        if choice == '1':
            try:
                num = float(input("Enter number: "))
                numbers.append(num)
                print(f"Added: {num}")
                print(f"Current numbers: {numbers}")

            except ValueError:
                print("Please enter a valid number!")

        elif choice == '2':
            if not numbers:
                print("No numbers to analyze!")
                continue

            total = sum(numbers)
            count = len(numbers)
            average = total / count
            maximum = max(numbers)
            minimum = min(numbers)

            print(f"\nStatistical Analysis:")
            print(f"Numbers: {numbers}")
            print(f"Count: {count}")
            print(f"Sum: {total:.2f}")
            print(f"Average: {average:.2f}")
            print(f"Maximum: {maximum:.2f}")
            print(f"Minimum: {minimum:.2f}")

        elif choice == '3':
            numbers.clear()
            print("Number list reset!")

        elif choice == '4':
            print("Goodbye!")
            break

        else:
            print("Invalid choice! Please enter 1-4.")

# Uncomment to run advanced version
# advanced_averager()

# Simple one-liner version using list comprehension
def quick_average():
    """
    Quick average calculator using list comprehension
    """
    print("Quick Average Calculator")
    input_str = input("Enter numbers separated by spaces: ")

    try:
        numbers = [float(x) for x in input_str.split()]

        if numbers:
            average = sum(numbers) / len(numbers)
            print(f"Numbers: {numbers}")
            print(f"Average: {average:.2f}")
        else:
            print("No numbers entered!")

    except ValueError:
        print("Please enter valid numbers separated by spaces!")

# Uncomment to run quick version
# quick_average()
```
