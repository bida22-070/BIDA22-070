```python
#Import libraries
from collections import deque
import time
import matplotlib.pyplot as plt
```

```python
# Exercise 1: Breadth-First Search (BFS)
def bfs(graph, start, goal):
    visited = set()
    queue = deque([(start, [])])

    while queue:
        node, path = queue.popleft()
        if node == goal:
            return path + [node]
        if node not in visited:
            visited.add(node)
            for neighbor in graph[node]:
                queue.append((neighbor, path + [node]))
    return None  # Goal not found

# Graph 1
graph1 = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

# Graph 2
graph2 = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F', 'G'],
    'D': ['B', 'H'],
    'E': ['B', 'H'],
    'F': ['C', 'H'],
    'G': ['C'],
    'H': ['D', 'E', 'F']
}

print("Exercise 1: BFS Algorithm Tests")

# Test different start and end points
test_cases = [
    ('A', 'E'),
    ('C', 'D'),
    ('F', 'B')
]

for i, (start, goal) in enumerate(test_cases, 1):
    start_time = time.time()
    result = bfs(graph1, start, goal)
    end_time = time.time()

    print(f"Test {i}: Start={start}, Goal={goal}")
    if result:
        print(f"Path found: {result}")
    else:
        print("No path found")
    print(f"Time taken: {end_time - start_time:.6f} seconds")
    print("-" * 40)

# Test with Graph 2
print("Testing Graph 2 ")
result_graph2 = bfs(graph2, 'A', 'H')
print(f"Path from A to H in Graph 2: {result_graph2}")
```

```python
# Exercise 2: Depth-First Search (DFS)
def dfs(graph, node, goal, visited=None):
    if visited is None:
        visited = set()

    if node == goal:
        return [node]

    if node not in visited:
        visited.add(node)
        for neighbor in graph[node]:
            path = dfs(graph, neighbor, goal, visited)
            if path:
                return [node] + path
    return None  # Goal not found

print("Exercise 2: DFS Algorithm Tests")
test_cases = [
    ('A', 'E'),
    ('C', 'D'),
    ('F', 'B')
]

for i, (start, goal) in enumerate(test_cases, 1):
    start_time = time.time()
    result = dfs(graph1, start, goal)
    end_time = time.time()

    print(f"Test {i}: Start={start}, Goal={goal}")
    if result:
        print(f"Path found: {result}")
    else:
        print("No path found")
    print(f"Time taken: {end_time - start_time:.6f} seconds")
    print("-" * 40)

# Compare BFS vs DFS
print("BFS vs DFS Comparison")
start_node, goal_node = 'A', 'F'

# BFS
bfs_start = time.time()
bfs_result = bfs(graph1, start_node, goal_node)
bfs_time = time.time() - bfs_start

# DFS
dfs_start = time.time()
dfs_result = dfs(graph1, start_node, goal_node)
dfs_time = time.time() - dfs_start

print(f"BFS Path: {bfs_result}, Time: {bfs_time:.6f}s")
print(f"DFS Path: {dfs_result}, Time: {dfs_time:.6f}s")
```

```python
# Exercise 3: Limited Depth DFS (LDS)
def dls(graph, node, goal, depth, visited=None):
    if visited is None:
        visited = set()

    if depth == 0:
        return None

    if node == goal:
        return [node]

    if node not in visited:
        visited.add(node)
        for neighbor in graph[node]:
            path = dls(graph, neighbor, goal, depth - 1, visited)
            if path:
                return [node] + path
    return None

print("Exercise 3: Limited Depth DFS Tests")
depths_to_test = [1, 2, 3, 4, 5]

for depth in depths_to_test:
    result = dls(graph1, 'A', 'F', depth)
    print(f"Depth {depth}: {result if result else 'No path found'}")

# Test breaking the algorithm
print("Testing Algorithm Limits")
result_impossible = dls(graph1, 'A', 'F', 1)
print(f"Depth 1 (should fail): {result_impossible if result_impossible else 'No path found - as expected'}")

# Create a larger graph for comparison
large_graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F', 'G'],
    'D': ['B', 'H', 'I'],
    'E': ['B', 'J', 'K'],
    'F': ['C', 'L', 'M'],
    'G': ['C', 'N'],
    'H': ['D', 'O'],
    'I': ['D'],
    'J': ['E'],
    'K': ['E', 'P'],
    'L': ['F'],
    'M': ['F'],
    'N': ['G'],
    'O': ['H'],
    'P': ['K']
}

print("Algorithm Comparison on Large Graph")
start_large, goal_large = 'A', 'P'

# BFS
bfs_start = time.time()
bfs_large = bfs(large_graph, start_large, goal_large)
bfs_large_time = time.time() - bfs_start

# DFS
dfs_start = time.time()
dfs_large = dfs(large_graph, start_large, goal_large)
dfs_large_time = time.time() - dfs_start

# LDS
lds_start = time.time()
lds_large = dls(large_graph, start_large, goal_large, 10)  # Increased depth
lds_large_time = time.time() - lds_start

print(f"BFS: {bfs_large}, Time: {bfs_large_time:.6f}s")
print(f"DFS: {dfs_large}, Time: {dfs_large_time:.6f}s")
print(f"LDS (depth=10): {lds_large}, Time: {lds_large_time:.6f}s")
```

```python
# Exercise 4: Maze Solving with BFS and DFS
def bfs_maze(maze, start, goal):
    rows, cols = len(maze), len(maze[0])
    visited = set()
    parent = {}  # To keep track of the parent of each cell in the path

    def is_valid(x, y):
        return (0 <= x < rows and 0 <= y < cols and
                maze[x][y] != '#' and (x, y) not in visited)

    queue = deque([start])
    visited.add(start)
    found = False

    while queue:
        x, y = queue.popleft()
        if (x, y) == goal:
            found = True
            break

        # Define possible moves: up, down, left, right
        moves = [(0, 1), (1, 0), (0, -1), (-1, 0)]
        for dx, dy in moves:
            new_x, new_y = x + dx, y + dy
            if is_valid(new_x, new_y):
                queue.append((new_x, new_y))
                visited.add((new_x, new_y))
                parent[(new_x, new_y)] = (x, y)

    if not found:
        return "No path found", maze, 0

    # Reconstruct the path from the goal to the start
    path = []
    x, y = goal
    while (x, y) != start:
        path.append((x, y))
        x, y = parent[(x, y)]
    path.append(start)
    path.reverse()

    # Create a copy of maze to mark path
    maze_copy = [row[:] for row in maze]
    for x, y in path:
        if maze_copy[x][y] != 'S' and maze_copy[x][y] != 'G':
            maze_copy[x][y] = '*'

    return "Path found", maze_copy, len(path)

def dfs_maze(maze, start, goal):
    rows, cols = len(maze), len(maze[0])
    visited = set()
    path = []  # Store the path

    def is_valid(x, y):
        return (0 <= x < rows and 0 <= y < cols and
                maze[x][y] != '#' and (x, y) not in visited)

    def dfs_search(x, y):
        if x == goal[0] and y == goal[1]:
            path.append((x, y))
            return True

        visited.add((x, y))
        path.append((x, y))

        # Define possible moves: up, down, left, right
        moves = [(0, 1), (1, 0), (0, -1), (-1, 0)]
        for dx, dy in moves:
            new_x, new_y = x + dx, y + dy
            if is_valid(new_x, new_y):
                if dfs_search(new_x, new_y):
                    return True

        path.pop()  # Remove the current cell from the path if no valid moves
        return False

    if dfs_search(start[0], start[1]):
        # Create a copy of maze to mark path
        maze_copy = [row[:] for row in maze]
        for x, y in path:
            if maze_copy[x][y] != 'S' and maze_copy[x][y] != 'G':
                maze_copy[x][y] = '*'
        return "Path found", maze_copy, len(path)
    else:
        return "No path found", maze, 0

# Define the maze
maze = [
    ['S', '.', '.', '.', '.', '.', '#', '.', '.'],
    ['#', '#', '#', '#', '#', '.', '#', '#', '#'],
    ['.', '.', '.', '.', '.', '.', '.', '.', 'G'],
    ['#', '#', '#', '#', '#', '#', '#', '#', '#']
]

# Starting and goal positions
start_position = (0, 0)
goal_position = (2, 8)

print("Exercise 4: Maze Solving Comparison")

# BFS Maze
print("Original Maze:")
for row in maze:
    print(' '.join(row))
print()

bfs_start = time.time()
bfs_result, bfs_maze_result, bfs_steps = bfs_maze(maze, start_position, goal_position)
bfs_time = time.time() - bfs_start

print("BFS Maze Solution:")
print(f"Result: {bfs_result}")
print(f"Steps: {bfs_steps}")
print(f"Time: {bfs_time:.6f}s")
for row in bfs_maze_result:
    print(' '.join(row))
print()

# DFS Maze
dfs_start = time.time()
dfs_result, dfs_maze_result, dfs_steps = dfs_maze(maze, start_position, goal_position)
dfs_time = time.time() - dfs_start

print("DFS Maze Solution:")
print(f"Result: {dfs_result}")
print(f"Steps: {dfs_steps}")
print(f"Time: {dfs_time:.6f}s")
for row in dfs_maze_result:
    print(' '.join(row))
```
